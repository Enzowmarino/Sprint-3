```python
import pyttsx3, speech_recognition as sr

engine = pyttsx3.init()

voices = engine.getProperty("voices")
engine.setProperty("voice", voices[0].id)
engine.setProperty("volume", 1)
engine.setProperty("rate", 155)

engine.runAndWait() 

def speak(audio):
    engine.say(audio)
    engine.runAndWait()

def command():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Ouvindo...")
        speak("Ouvindo!")
        r.adjust_for_ambient_noise = 1.25
        r.pause_threshold = 1
        audio = r.listen(source)
    try:
        print("Reconhecendo...")
        speak("Espera um pouco pra eu tentar entender.")
        query = r.recognize_google(audio, language="pt")
    except:
        return "©empty_^_^_queryª"
    return query

speak("Olá, sou o seu assistente de Compras! Posso te ajudar em algo?  Por favor, responda com me ajude ou com não.")

if __name__ == "__main__":
        
    while True:
        query = command().lower()
        if "me ajude" in query or "me ajuda" in query:
            print("Ajuda")
            speak("Tenho algumas opções em que posso te ajudar, como primeira, achar um produto ou segunda, verificar o preço de um produto. Escolha uma das duas opções.")  
            while True:
                query = command().lower()   
                    
                if "primeira" in query or "primeiro" in query:
                    
                    
                    print("Escolha: 1")
                    speak("Você escolheu a opção número 1.")
                    speak("Tenho algumas opções de produtos para te ajudar a encontrar, como frutas, carnes e laticínios. Escolha um deles! Ou se quiser escolher outra ajuda, diga sair!")
                    
                    while True:
                        
                        query = command().lower()
                        
                        if "frutas" in query or "fruta" in query:
                            speak("As frutas estão no corredor 5, próximo as geladeiras de frios, na prateleira inferior.")
                        elif "carnes" in query or "carne" in query:
                            speak("As carnes podem ser encontradas, no último corredor do mercado, de número 8 na seção de açogue.")
                        elif "laticinio" in query or "laticínios" in query:
                            speak("Os laticínios estão no início do corredor 7, próximos das bebidas.") 
                    
                        
                        elif "fim" in query or "sair" in query : 
                            print("sair")
                            speak("Voltando ao Menu das opções de ajuda")
                            break
                    
                    
                        else:
                            print("Opção inválida.")
                            speak("não entendi o que você disse.")
                            speak("Por favor, vamos tentar novamente. escolha uma opção válida.")
                
                elif "segunda" in query or "segundo" in query:
                    print("Escolha: 2")
                    speak("você escolheu a opção numero 2")
                    speak("Bem, por enquanto, posso te dizer o preço de alguns produtos, como o requeijão, café, ou leite, escolha um!")
                    while True:
                        
                        query = command().lower()
                        
                        if "requeijão" in query or "requeijao" in query:
                            speak("o preço do requeijão é de 7 reais o pode pequeno e 15 reais o pote grande")
                        elif "café" in query or "cafe" in query or "CAFE" in query:
                            speak("O preço do café é de 15 reais o quilo.")
                        elif "leite" in query or "LEITE" in query or "Leites" in query:
                            speak("O litro de leite varia de 4 a 5 reais dependendo da marca.") 
                    
                        
                        elif "fim" in query or "sair" in query : 
                            print("sair")
                            speak("Voltando ao Menu das opções de ajuda")
                            break
                    
                    
                        else:
                            print("Opção inválida.")
                            speak("Não entendi o que você disse.")
                            speak("Por favor, vamos tentar novamente. Escolha uma opção válida.")
                
                elif "fim" in query or "sair" in query : 
                    print("sair")
                    speak("Voltando ao menu do sim e não.")
                    break
                            
                else:
                    print("Opção inválida.")
                    speak("Não entendi o que você disse.")
                    speak("Por favor, vamos tentar novamente. Escolha uma opção válida.")
            
        elif "não" in query:
            print("Opção: 2")
            speak("Até logo!")
            break
                
                  
        else:
            print("Opção inválida.")
            speak("Não entendi o que você disse.")
            speak("Por favor, vamos tentar novamente. escolha uma opção válida.")
            
        
        
```
